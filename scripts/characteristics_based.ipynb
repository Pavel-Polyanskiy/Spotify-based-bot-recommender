```python
df = pd.read_csv(cb_data_path, index_col=0)
df['clust'] = df['clust'].astype('float32')
```

```python
df1 = df
df1 = df1.drop('genre1', axis= 1)
df1 = df1.drop('genre2', axis= 1)
df1 = df1.drop('cluster_means', axis= 1)
#df1 = df1.drop('clust', axis= 1)
df1 = df1.drop('song_name', axis= 1)
df1 = df1.drop('artist', axis= 1)
df1 = df1.drop('track_id', axis= 1)
df1 = df1.drop('genre', axis= 1)
df1 = df1.drop('p', axis= 1)
```

```python
df1.index = df['track_id']
df1['danceability'] = df1['danceability'].astype('float32')
df1['energy'] = df1['energy'].astype('float32')
df1['key'] = df1['key'].astype('float32')
df1['loudness'] = df1['loudness'].astype('float32')
df1['speechiness'] = df1['speechiness'].astype('float32')
df1['acousticness'] = df1['acousticness'].astype('float32')
df1['instrumentalness'] = df1['instrumentalness'].astype('float32')
df1['liveness'] = df1['liveness'].astype('float32')
df1['valence'] = df1['valence'].astype('float32')
df1['tempo'] = df1['tempo'].astype('float32')
```

```python
def give_recommendations_clus(x, y, num): # x - artist, y - song name, num - number of recs
    index = df[df['artist'].str.contains(x)]
    index = index[index['song_name'].str.contains(y)]
    index = index.drop_duplicates(subset='artist',keep='first')
    clus = index['clust'].iloc[0]
    spd = index['song_name'].iloc[0]
    index = np.array(index['track_id'])
    index = index.tolist()
    index_recomm =cos_sim.loc[:,index]
    index_recomm = index_recomm.sort_values(index_recomm.columns[0], ascending = False)
    index_recomm['track_id'] = index_recomm.index
    index_recomm.index = np.arange (len(index_recomm))
    clus = df[df['clust'].str.contains(clus)]
    clus = clus[['track_id']]
    song_rec = df [['track_id', 'artist', 'song_name', 'genre1', 'genre2']]
    song_rec = song_rec.merge(index_recomm, how = "right")
    song_rec = song_rec.merge(clus, how = "right")
    song_rec= song_rec.drop(song_rec[song_rec['song_name'] == spd].index)
    k = len (index_recomm)
    song_rec = song_rec.head(k)
    song_rec.index = np.arange (len(song_rec))
    song_rec = song_rec.iloc[permutation(song_rec.index)]
    h = num*2
    song_rec = song_rec.head(h)
    song_rec = song_rec.sort_values(song_rec.columns[5], ascending = False)
    song_rec = song_rec [['artist', 'song_name', 'genre1', 'track_id']]
    song_rec.index = np.arange (len(song_rec))
    song_rec = song_rec.head(num)
    return song_rec
```

```python
def give_recommendations_genre(x, num):
    genre = df[df['genre'].str.contains(x)]
    genre = genre[['track_id']]
    index = 'silence'
    index_recomm = pd.DataFrame(cos_sim.loc[:,index])
    index_recomm = index_recomm.sort_values(index_recomm.columns[0], ascending = True)
    index_recomm['track_id'] = index_recomm.index
    index_recomm.index = np.arange (len(index_recomm))
    song_rec = df [['track_id', 'artist', 'song_name', 'genre1', 'genre2']]
    song_rec = song_rec.merge(index_recomm, how = "right")
    song_rec = song_rec.merge(genre, how = "right")
    song_rec = song_rec[['artist', 'song_name', 'genre1', 'genre2']]
    song_rec = song_rec.iloc[permutation(song_rec.index)]
    song_rec.index = np.arange (len(song_rec))
    song_rec = song_rec.head(num)
    return song_rec
```

```python
def get_recommendation_by_characteristics(track_id, n):
    artist = id2track[id2track.index == track_id]['artist'][0]
    song_name = id2track[id2track.index == track_id]['song_name'][0]

    recommendation_list = list(cos_sim.loc[track_id, :].sort_values(ascending = False)[:n].index)
    track_names = [f'{id2track.loc[rec][1]} by {id2track.loc[rec][0]}' for rec in recommendation_list]
    track_url_list = [f'https://open.spotify.com/track/{rec}' for rec in recommendation_list]

    return track_names, track_url_list
```
